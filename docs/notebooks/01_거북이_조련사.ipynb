## 퀴즈 1: forward() 이해하기

````python
#@title forward(100)에서 100은 무엇을 의미하나요?
answer = "\uD53D\uC140 (\uD654\uBA74\uC758 \uC810)" #@param ["\uD53D\uC140 (\uD654\uBA74\uC758 \uC810)", "\uCD08 (\uC2DC\uAC04)", "\uBC88 (\uBC18\uBCF5 \uD69F\uC218)"]

if answer == "픽셀 (화면의 점)":
    print("정답입니다! forward(100)은 100 픽셀만큼 앞으로 이동합니다.")
else:
    print("다시 생각해보세요. forward()의 숫자는 이동 거리를 픽셀 단위로 나타냅니다.")
````

## 퀴즈 2: 회전 방향

````python
#@title right(90)은 어느 방향으로 회전하나요?
answer = "\uC624\uB978\uCABD (\uC2DC\uACC4 \uBC29\uD5A5)" #@param ["\uC624\uB978\uCABD (\uC2DC\uACC4 \uBC29\uD5A5)", "\uC67C\uCABD (\uBC18\uC2DC\uACC4 \uBC29\uD5A5)", "\uC55E\uC73C\uB85C"]

if answer == "오른쪽 (시계 방향)":
    print("정답! right()은 시계 방향(오른쪽)으로 회전합니다.")
else:
    print("다시 생각해보세요. right = 오른쪽 = 시계 방향입니다.")
````
