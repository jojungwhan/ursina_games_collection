## 퀴즈 1: 변수 할당

````python
#@title 변수에 값을 넣을 때 사용하는 기호는?
answer = "=" #@param ["==", "=", ":="]

if answer == "=":
    print("정답! = 기호로 변수에 값을 저장합니다.")
    print("예: size = 100")
else:
    print("다시 생각해보세요.")
    print("== 는 비교할 때, = 는 저장할 때 사용합니다.")
````

## 연습 3: 배경색 바꾸기 (bgcolor)

ColabTurtle에서는 `bgcolor()` 대신 `initializeTurtle()`의 설정을 사용합니다.

배경색을 바꿔보세요!

````python
#@title 배경색 실험
background = "black" #@param ["white", "black", "yellow", "lightblue", "pink"]

from ColabTurtle.Turtle import *
initializeTurtle(initial_window_color=background)

# 배경색이 바뀌었나요?
forward(100)
right(90)
forward(100)
````

## 연습 4: 펜 색상 바꾸기 (pencolor)

`color()` 함수로 펜 색상을 바꿀 수 있습니다.

**힌트**: `color('red')`, `color('blue')`, `color('green')`

````python
# 연습 4: 펜 색상 바꾸기
from ColabTurtle.Turtle import *
initializeTurtle(initial_window_color="black")

# Step 1: 펜 색상을 빨간색으로 바꾸세요
# 힌트: color('red')


# Step 2: 앞으로 100 이동
forward(100)

# Step 3: 펜 색상을 파란색으로 바꾸세요


# Step 4: 오른쪽 90도 회전 후 앞으로 100 이동
right(90)
forward(100)
````

## 퀴즈 2: 문자열

````python
#@title 색상 이름을 저장할 때 어떻게 작성하나요?
answer = "my_color = 'red'" #@param ["my_color = red", "my_color = 'red'", "my_color : red"]

if answer == "my_color = 'red'":
    print("정답! 문자열(글자)은 따옴표로 감싸야 합니다.")
else:
    print("다시 생각해보세요.")
    print("글자(문자열)는 따옴표(' 또는 \")로 감싸야 합니다.")
````

## 연습 6: 변수 재사용하기

같은 변수를 여러 번 사용하면 **일관성**있게 값을 바꿀 수 있습니다!

````python
#@title 변수 재사용 실험
size = 80 #@param {type:"slider", min:20, max:150, step:10}
my_color = "orange" #@param ["red", "orange", "yellow", "green", "blue", "purple"]

from ColabTurtle.Turtle import *
initializeTurtle(initial_window_color="black")

color(my_color)

# size 변수를 여러 번 사용!
forward(size)
right(90)
forward(size)
right(90)
forward(size)
right(90)
forward(size)

print(f"사각형 크기: {size}, 색상: {my_color}")
````
