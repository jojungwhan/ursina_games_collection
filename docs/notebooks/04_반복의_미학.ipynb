## 연습 2: 들여쓰기 (Indentation)

**들여쓰기**는 `for` 반복문에서 **매우 중요**합니다!

- 들여쓰기된 코드 = 반복되는 코드
- 들여쓰기 안 된 코드 = 반복 안 됨

**공백 4칸** 또는 **Tab 1번**으로 들여쓰기합니다.

````python
# 들여쓰기 비교

print("=== 들여쓰기 O ===")
for i in range(3):
    print("반복!")    # 들여쓰기 있음 - 3번 출력

print("\n=== 들여쓰기 X ===")
for i in range(3):
    pass  # 빈 반복
print("한 번만!")     # 들여쓰기 없음 - 1번만 출력
````

## 퀴즈 2: 반복 횟수

````python
#@title 오각형을 그리려면 몇 번 반복해야 할까요?
answer = "5\uBC88" #@param ["4\uBC88", "5\uBC88", "6\uBC88"]

if answer == "5번":
    print("정답! 오각형은 5개의 변이 있으므로 5번 반복합니다.")
    print("회전 각도는 360/5 = 72도입니다.")
else:
    print("다시 생각해보세요.")
    print("n각형을 그리려면 n번 반복합니다.")
````

````python
#@title 다각형 실험
sides = 6 #@param {type:"slider", min:3, max:12, step:1}

from ColabTurtle.Turtle import *
initializeTurtle()

angle = 360 / sides  # 회전 각도 계산

for i in range(sides):
    forward(50)
    right(angle)

print("다각형 완성!")
print("변의 개수:")
print(sides)
print("회전 각도:")
print(angle)
````

## 연습 7: penup/pendown

- `penup()` - 펜을 들어서 이동해도 선이 안 그려짐
- `pendown()` - 펜을 내려서 다시 선이 그려짐

````python
# 연습 7: penup/pendown
from ColabTurtle.Turtle import *
initializeTurtle()

# 선 그리기
forward(50)

# 펜 들기 (이동해도 선 안 그려짐)
penup()
forward(30)

# 펜 내리기 (다시 선 그려짐)
pendown()
forward(50)

print("점선처럼 보이죠?")
````

````python
# 점선 그리기
from ColabTurtle.Turtle import *
initializeTurtle()

# 점선 패턴을 5번 반복
for i in range(5):
    # 아래에 점선 패턴을 타이핑하세요!
    # pendown() -> forward(20) -> penup() -> forward(10)
    pass  # 이 줄을 지우고 코드를 작성하세요
````

## 도전 과제: 여러 사각형 그리기

반복문을 중첩해서 여러 사각형을 그려보세요!

````python
# 도전 과제: 회전하는 사각형들
from ColabTurtle.Turtle import *
initializeTurtle(initial_window_color="black")
color('cyan')

# 12개의 사각형을 30도씩 회전하며 그리기
for j in range(12):           # 12번 반복
    for i in range(4):        # 사각형 (4번 반복)
        forward(80)
        right(90)
    right(30)                 # 30도 회전
````
