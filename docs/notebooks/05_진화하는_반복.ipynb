## 연습 1: 반복 변수 i란?

`for i in range(5):`에서 `i`는 **반복할 때마다 바뀌는 변수**입니다.

| 반복 | i 값 |
|:---:|:---:|
| 1번째 | 0 |
| 2번째 | 1 |
| 3번째 | 2 |
| 4번째 | 3 |
| 5번째 | 4 |

```python
# 연습 1: i 값 확인하기

print("i 값 변화:")
for i in range(5):
    print(f"  i = {i}")
```

```python
# 직접 타이핑해보세요!
# range(10)으로 i 값 출력하기

print("0부터 9까지:")
# 아래에 코드를 타이핑하세요
```

## 연습 2: forward(i) - 점점 길어지는 선

`forward(i)`를 사용하면 **점점 길어지는 선**을 그릴 수 있습니다!

```python
# 연습 2: forward(i)
from ColabTurtle.Turtle import *
initializeTurtle()

for i in range(10):
    forward(i * 5)  # 0, 5, 10, 15, ... 45
    right(90)

print("점점 커지는 사각형 나선!")
```

## 연습 3: i를 계산에 사용하기

`i`를 다양한 계산에 사용할 수 있습니다!

- `i * 10` - 10배
- `i + 5` - 5 더하기
- `i * i` - 제곱

```python
# i 계산 예시

print("i * 10:")
for i in range(5):
    print(f"  i={i} -> {i * 10}")

print("\ni + 50:")
for i in range(5):
    print(f"  i={i} -> {i + 50}")
```

## 연습 4: circle(i) - 점점 커지는 원

ColabTurtle에서는 `circle(반지름)`으로 원을 그립니다.

```python
# 연습 4: 점점 커지는 원
from ColabTurtle.Turtle import *
initializeTurtle()

for i in range(1, 6):  # 1, 2, 3, 4, 5
    circle(i * 20)     # 20, 40, 60, 80, 100
```

```python
#@title 원 크기 실험
circles = 8 #@param {type:"slider", min:3, max:15, step:1}
multiplier = 15 #@param {type:"slider", min:5, max:30, step:5}

from ColabTurtle.Turtle import *
initializeTurtle(initial_window_color="black")
color('cyan')

for i in range(1, circles + 1):
    circle(i * multiplier)

print(f"{circles}개의 원, 간격: {multiplier}")
```

## 연습 5: goto(x, y) - 위치 이동

`goto(x, y)`는 거북이를 **(x, y)** 좌표로 이동시킵니다.

- 화면 중앙이 (0, 0)
- x가 양수: 오른쪽
- y가 양수: 위쪽

```python
# 연습 5: goto로 점 찍기
from ColabTurtle.Turtle import *
initializeTurtle()

penup()  # 선 안 그리기

# 5개의 점을 일정 간격으로 찍기
for i in range(5):
    goto(i * 50, 200)  # x = 0, 50, 100, 150, 200
    pendown()
    circle(5)  # 작은 원으로 점 표시
    penup()
```

## 연습 6: i를 x, y 좌표에 사용

i를 계산해서 다양한 위치에 도형을 그려봅시다!

```python
# 연습 6: 대각선 위치에 원 그리기
from ColabTurtle.Turtle import *
initializeTurtle(initial_window_color="black")
color('yellow')
penup()

for i in range(5):
    # x와 y 모두 i에 따라 변함
    x = i * 40
    y = 400 - i * 40  # 위에서 아래로
    goto(x, y)
    pendown()
    circle(15)
    penup()

print("대각선으로 원이 그려졌어요!")
```

## 도전 과제: 무지개 원

i를 활용해서 점점 커지는 무지개 원을 그려보세요!

```python
# 도전 과제: 무지개 원
from ColabTurtle.Turtle import *
initializeTurtle(initial_window_color="black")

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple']

for i in range(6):
    color(colors[i])  # i번째 색상 사용
    circle((i + 1) * 20)  # 20, 40, 60, 80, 100, 120

print("무지개 원 완성!")
```
